### 14. Apply LabelEncoder, OneHotEncoder, and OrdinalEncoder to a sample dataset and compare the resulting feature space dimensions.


```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
import seaborn as sns
```

```python
sample_data=({'color':['red','green','blue','green']})
```

```python
df=pd.DataFrame(sample_data)
```

```python
OH = OneHotEncoder(sparse_output=False)
LE = LabelEncoder()
OE = OrdinalEncoder()
```

```python
Hot = pd.DataFrame(OH.fit_transform(df[['color']]))
```

```python
Hot
```

```python
df1=df.join(Hot)
```

```python
df1
```

```python
Label = pd.DataFrame(LE.fit_transform(df.color))
```

```python
Label
```

```python
df2=df.join(Label)
```

```python
df2
```

```python
Ordinal = pd.DataFrame(OE.fit_transform(df[['color']]))
```

```python
df3=df.join(Ordinal)
```

```python
df3
```

Dimension of One Hot Encoding > Label Encoding = Ordinal Encoding

### 15. Apply both StandardScaler and MinMaxScaler on any numerical dataset and visualize the difference using boxplots. 

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
scaler1=StandardScaler()
scaler2=MinMaxScaler()
```

```python
df=pd.DataFrame(sns.load_dataset('titanic'))
```

```python
df.head()
```

```python
fare_scaled1 = pd.DataFrame(scaler1.fit_transform(df[['fare']]))
fare_scaled2 = pd.DataFrame(scaler2.fit_transform(df[['fare']]))
fare_scaled=pd.merge(fare_scaled1,fare_scaled2)
```

```python
fare_scaled1['key']=range(len(fare_scaled1))
fare_scaled2['key']=range(len(fare_scaled2))
```
