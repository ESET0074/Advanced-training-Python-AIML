```python
from sklearn.model_selection import train_test_split
y = housing.Price
x = housing.drop(['Price'], axis=1)
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(x, y, train_size=0.8, random_state=0)
```

```python
x=x.dropna(axis=0)
```

```python
low_cardinality_cols = [cname for cname in X_train.columns if X_train[cname].nunique() < 10 and X_train[cname].dtype == 'object']
```

```python
numerical_cols =[cname for cname in X_train.columns if X_train[cname].dtype in ['int64','float64']]
```

```python
total_cols = low_cardinality_cols + numerical_cols
```

```python
new_X_train = X_train[total_cols].copy()
```

```python
new_X_valid = X_valid[total_cols].copy()
```

```python
cat = (new_X_train.dtypes == 'object')
list_cat = list(cat[cat].index)
```

```python
OH_encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
OH_cols_train = pd.DataFrame(OH_encoder.fit_transform(new_X_train[list_cat]))
OH_cols_valid = pd.DataFrame(OH_encoder.transform(new_X_valid[list_cat]))
```

```python
OH_cols_train.index =new_X_train.index
OH_cols_valid.index =new_X_valid.index

num_x_train = new_X_train.drop(list_cat, axis = 1)
num_x_valid = new_X_valid.drop(list_cat, axis = 1)

OH_x_train=pd.concat([num_x_train, OH_cols_train], axis =1)
OH_x_valid=pd.concat([num_x_valid, OH_cols_train], axis =1)

OH_x_train.columns = OH_x_train.columns.astype(str)
OH_x_valid.columns = OH_x_valid.columns.astype(str)
```

#### 2. Data Imputation

```python
car['Engine Fuel Type']=car.groupby('Model')['Engine Fuel Type'].transform(lambda x:x.fillna(x.mode().iloc[0] if not x.mode().empty else None))
```

```python
car['Number of Doors']=car['Number of Doors'].fillna(car['Number of Doors'].mean())
```
